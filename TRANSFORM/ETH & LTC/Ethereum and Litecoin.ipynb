```python
import os
import csv
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
Ethereum_df= pd.read_csv('coin_Ethereum.csv')
Ethereum_df.head()
```

```python
len(Ethereum_df)
```

```python
Litecoin_df= pd.read_csv('coin_Litecoin.csv')
Litecoin_df.head()
```

```python
len(Litecoin_df)
```

```python
Ethereum19to20_df = Ethereum_df[(Ethereum_df['Date'] >"2019-01-01") & (Ethereum_df['Date'] <="2020-12-31")]
Ethereum19to20_df.head(10)
```

```python
#Find average volume for Ethereum
Ethereum19to20_df[["Volume"]].mean()
```

```python
Ethereum_filter_df=Ethereum19to20_df[Ethereum19to20_df['Volume'] >= 1.060659e+10]
Ethereum_filter_df.head()
```

```python
len(Ethereum19to20_df)
```

```python
Litecoin19to20_df = Litecoin_df[(Litecoin_df['Date'] > "2019-01-01") & (Litecoin_df['Date'] <= "2020-12-31")]
Litecoin19to20_df.head(10)
```

```python
#Find average volume of Litecoin
Litecoin19to20_df[["Volume"]].mean()
```

```python
Litecoin_filter_df=Litecoin19to20_df[Litecoin19to20_df['Volume'] >= 3.238216e+09]
Litecoin_filter_df.head()
```

```python
len(Litecoin19to20_df)
```

```python
Ethereum_filter_df.to_csv("Ethereum19to20_df.csv", index = False, header=True)
```

```python
Litecoin_filter_df.to_csv("Litecoin19to20_df.csv", index = False, header=True)
```
